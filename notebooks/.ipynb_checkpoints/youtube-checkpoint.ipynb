```python
df['date'] = pd.to_datetime(df.index, format='%Y-%m-%d')
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
```

```python
df.columns
```

```python
df['dev_total'] = df[['Computer', 'Game console', 'Mobile phone', 'TV', 'Tablet']].sum(axis=1)
```

```python
df['geo_total'] = df[['AR', 'AT',
       'AU', 'BD', 'BE', 'BG', 'BR', 'BY', 'CA', 'CH', 'CL', 'CN', 'CO', 'CR',
       'CZ', 'DE', 'DK', 'DZ', 'EC', 'EG', 'ES', 'FI', 'FR', 'GB', 'GR', 'HU',
       'ID', 'IE', 'IL', 'IN', 'IT', 'JP', 'KR', 'KZ', 'LT', 'LU', 'MO', 'MX',
       'MY', 'NL', 'NO', 'NZ', 'PE', 'PH', 'PL', 'PT', 'PY', 'RO', 'RS', 'RU',
       'SA', 'SE', 'SG', 'SK', 'TH', 'TR', 'TW', 'UA', 'US', 'VE', 'VN']].sum(axis=1)
```

```python
df['other_device'] = df['Views'] - df['dev_total']
df['other_country'] = df['Views'] - df['geo_total']
```

```python
df['dev_total_final'] = df[['Computer', 'Game console', 'Mobile phone', 'TV', 'Tablet', 'other_device']].sum(axis=1)
```

```python
df['geo_total_final'] = df[['AR', 'AT',
       'AU', 'BD', 'BE', 'BG', 'BR', 'BY', 'CA', 'CH', 'CL', 'CN', 'CO', 'CR',
       'CZ', 'DE', 'DK', 'DZ', 'EC', 'EG', 'ES', 'FI', 'FR', 'GB', 'GR', 'HU',
       'ID', 'IE', 'IL', 'IN', 'IT', 'JP', 'KR', 'KZ', 'LT', 'LU', 'MO', 'MX',
       'MY', 'NL', 'NO', 'NZ', 'PE', 'PH', 'PL', 'PT', 'PY', 'RO', 'RS', 'RU',
       'SA', 'SE', 'SG', 'SK', 'TH', 'TR', 'TW', 'UA', 'US', 'VE', 'VN', 'other_country']].sum(axis=1)
```

```python
df.drop(['dev_total', 'geo_total', 'dev_total_final', 'geo_total_final'], axis=1, inplace=True) 
```

```python
df.columns
```

```python
devices = ['Computer', 'Game console', 'Mobile phone', 'TV', 'Tablet', 'other_device']

for device in devices:
    df['{device}_percent'.format(device=device)] = df[device]/df.Views
```

```python
countries = ['AR', 'AT',
       'AU', 'BD', 'BE', 'BG', 'BR', 'BY', 'CA', 'CH', 'CL', 'CN', 'CO', 'CR',
       'CZ', 'DE', 'DK', 'DZ', 'EC', 'EG', 'ES', 'FI', 'FR', 'GB', 'GR', 'HU',
       'ID', 'IE', 'IL', 'IN', 'IT', 'JP', 'KR', 'KZ', 'LT', 'LU', 'MO', 'MX',
       'MY', 'NL', 'NO', 'NZ', 'PE', 'PH', 'PL', 'PT', 'PY', 'RO', 'RS', 'RU',
       'SA', 'SE', 'SG', 'SK', 'TH', 'TR', 'TW', 'UA', 'US', 'VE', 'VN', 'other_country']

for country in countries:
    df['{country}_percent'.format(country=country)] = df[country]/df.Views
```

```python
df['north_america_total'] = df[['CA', 'CR', 'MX', 'US']].sum(axis=1)
df['south_america_total'] = df[['AR', 'BR', 'CL', 'CO', 'EC', 'PE', 'PY', 'VE']].sum(axis=1)
df['europe_total'] = df[['AT', 'BE', 'BG', 'BY', 'CH', 'CZ', 'DE', 'DK', 'ES', 'FI', 'FR',
                        'GB', 'GR','HU', 'IE', 'IT', 'LT', 'LU', 'NL', 'PL', 'PT', 'RO',
                        'RS', 'RU', 'SE', 'SK','UA']].sum(axis=1)
df['australasia_total'] = df[['AU', 'NZ']].sum(axis=1)
df['asia_total'] = df[['BD', 'CN', 'ID', 'IL', 'IN', 'JP', 'KR', 'KZ', 'MO', 'MY', 'PH',
                      'SA', 'SG','TH', 'TW', 'VN', 'TR']].sum(axis=1)
df['africa_total'] = df[['DZ', 'EG']].sum(axis=1)
```

```python
continents = ['north_america_total', 'south_america_total', 'europe_total',
       'australasia_total', 'asia_total', 'africa_total']

for continent in continents:
    df['{continent}_percent'.format(continent=continent)] = df[continent]/df.Views
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df = df.loc[df.year >= 2020].copy()
```
