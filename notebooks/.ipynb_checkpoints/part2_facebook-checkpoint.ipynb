#### Import Packages & Functions

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import re
import joblib
from os import path
from PIL import Image

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import WordPunctTokenizer

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn import set_config

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

def stem_sentences(sentence):
    """
    This function takes a full sentence, splits it into words, 
    stems using the stemmer created outside of the function and joins them back together
    into a sentence format.
    """
    stemmer = PorterStemmer()
    tokens = sentence.split()
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return ' '.join(stemmed_tokens)

def model_fitter(model, X_train, y_train, X_test, y_test):
    """
    This function takes a model and the X, y variables.
    It then fits the model and returns R2 scores for train & test sets.
    """
    model.fit(X_train, y_train)
    cv_scores = cross_val_score(model, X_train, y_train, cv=5)
    print("Training Cross-validated R2:", cv_scores)
    print("Training Mean cross-validated R2:", cv_scores.mean())
    print("Training R2 Score:", model.score(X_train, y_train))
    print("Test R2 Score:", model.score(X_test, y_test))

plt.style.use('ggplot')

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

#### EDA & Feature Extraction

We then clean up the remaining columns by dropping or filling in the null values as required. There were several columns where the null value simply meant 0.

```python
values = {'Caption type': 'Not available', 'Link clicks': 0.0, 'Other clicks': 0.0, 'Photo views': 0.0}
fb_data.fillna(value=values, inplace=True)
```

```python
fb_data.dropna(subset=['Title', 'Description'], axis=0, inplace=True)
```

We also change the types of some columns for better usability in the future.

```python
fb_data['Link clicks'] = fb_data['Link clicks'].astype(int)
fb_data['Other clicks'] = fb_data['Other clicks'].astype(int)
fb_data['Photo views'] = fb_data['Photo views'].astype(int)
```

Unfortunately the hour/minute of the `Publish time` column are not clear. There is no indication of what timezone it is referring to and there is no obvious correlation to the publish time listed on the profile vs the data downloaded. Because of this, we ignore it and focus on day, month, and year.

```python
fb_data['Publish time'] = pd.to_datetime(fb_data['Publish time'])
```

```python
fb_data['day_of_week'] = fb_data['Publish time'].dt.day_name()
fb_data['month'] = fb_data['Publish time'].dt.month
fb_data['year'] = fb_data['Publish time'].dt.year
```

```python
fb_data.drop(['Publish time'], axis=1, inplace=True)
```

### Modelling

We set our X & y variables.

Here we focus on the datetime variables.

```python
X = df[['day_of_week', 'month', 'year', 'Link clicks']].copy()
y = X.pop('Link clicks')
```

This is the average number of link clicks.

```python
y.mean()
```

We use the `train_test_split` function and follow with the similar `ColumnTransformer` and `Pipeline` steps as used in Part 1 of this project.

This time we have a regression problem so we begin with a default `LinearRegression()`.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
preprocessing = ColumnTransformer([('cat', OneHotEncoder(), ['day_of_week', 'month', 'year'])],
                                 remainder='passthrough')

pipe_linreg_date = Pipeline(steps=[('preprocessing', preprocessing),
                          ('model', LinearRegression())])
```

```python
set_config(display='diagram')
pipe_linreg_date
```

We are now working a regression problem so the scoring has been changed to R2 which tells us how much of the variation in link clicks can be explained by our feature variables. We want this to be close to 1.

```python
model_fitter(pipe_linreg_date, X_train, y_train, X_test, y_test)
```

Our scores our no where near 1. This model does not work.

Let's try our text variables.

```python
X = df[['stemmed_title', 'Link clicks']].copy()
y = X.pop('Link clicks')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
preprocessing = ColumnTransformer([
                               ('words', CountVectorizer(stop_words=stop), 'stemmed_title')],
                                 remainder='passthrough')

pipe_linreg_title = Pipeline(steps=[('preprocessing', preprocessing),
                          ('model', LinearRegression())])
```

```python
model_fitter(pipe_linreg_title, X_train, y_train, X_test, y_test)
```

```python
X = df[['stemmed_desc', 'Link clicks']].copy()
y = X.pop('Link clicks')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
preprocessing = ColumnTransformer([
                               ('words', CountVectorizer(stop_words=stop), 'stemmed_desc')],
                                 remainder='passthrough')

pipe_linreg_desc = Pipeline(steps=[('preprocessing', preprocessing),
                          ('model', LinearRegression())])
```

```python
model_fitter(pipe_linreg_desc, X_train, y_train, X_test, y_test)
```

None of the models seem to be working.

### Conclusion

While our modelling was unsuccessful, it is interesting to see that the same NLP approach used for the newsletter does not work on Facebook.

Our assumption that the text (to indicate subject matter) would influence how many times the link is clicked by a user is incorrect.

This shows a difference in how the information is consumed on either platform, and implies there is something more powerful at work here - perhaps the Facebook algorithm? Or current events and trending topics?
